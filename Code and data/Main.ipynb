### Analysis-1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# General plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Helper function for adding titles
def plot_title(ax, title):
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel(ax.get_xlabel(), fontsize=12)
    ax.set_ylabel(ax.get_ylabel(), fontsize=12)

# 1. Age Distribution

plt.figure()
ax = sns.histplot(data=df, x="Age", bins=20, kde=True, color="skyblue")
plot_title(ax, "Age Distribution Across Sectors")
plt.show()

# 2. Gender Distribution by Sector

plt.figure()
ax = sns.countplot(data=df, x="Sector", hue="Sex", palette="Set2")
plot_title(ax, "Gender Distribution by Sector")
plt.legend(title="Gender")
plt.show()

# 3. Household Size by Sector
plt.figure()
ax = sns.boxplot(data=df, x="Sector", y="Household Size", palette="Set3")
plot_title(ax, "Household Size Distribution by Sector")
plt.show()

# 4. Worker Participation Rate by Sector
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Worker Participation Ratio (CWS)", ci=None, palette="muted")
plot_title(ax, "Worker Participation Rate by Sector")
plt.show()

# 5. Labor Force Participation Rate by Gender and Sector
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Labour Force Participation Ratio (CWS)", hue="Sex", ci=None, palette="pastel")
plot_title(ax, "Labour Force Participation Rate by Gender and Sector")
plt.legend(title="Gender")
plt.show()

# 6. Monthly Household Expenditure by Social Group
plt.figure()
ax = sns.boxplot(data=df, x="Social Group", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="coolwarm")
plot_title(ax, "Monthly Household Expenditure by Social Group")
plt.xticks(rotation=45)
plt.show()

# 7. Household Expenditure by Sector
plt.figure()
ax = sns.boxplot(data=df, x="Sector", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="husl")
plot_title(ax, "Monthly Household Expenditure by Sector")
plt.show()

# 8. Household Expenditure by Household Size
plt.figure()
ax = sns.scatterplot(data=df, x="Household Size", y="Household's Usual Consumer Expenditure In A Month (Rs.)", alpha=0.6)
plot_title(ax, "Household Expenditure by Household Size")
plt.show()

# 9. Temporal Analysis of Worker Participation Rate
plt.figure()
ax = sns.lineplot(data=df, x="Month of Survey", y="Worker Participation Ratio (CWS)", hue="Sector", marker="o", palette="Dark2")
plot_title(ax, "Worker Participation Rate Over Time")
plt.xticks(rotation=45)
plt.show()

# 10. Correlation Between Variables
plt.figure()
corr_cols = [
    "Household Size", 
    "Household's Usual Consumer Expenditure In A Month (Rs.)", 
    "Worker Participation Ratio (CWS)", 
    "Labour Force Participation Ratio (CWS)"
]

correlation_matrix = df[corr_cols].corr()
ax = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plot_title(ax, "Correlation Between Key Metrics")
plt.show()

# 11. Gender Disparities in Participation Rate
plt.figure()
ax = sns.boxplot(data=df, x="Sex", y="Worker Participation Ratio (CWS)", palette="viridis")
plot_title(ax, "Gender Disparities in Worker Participation Rate")
plt.show()

# 12. Disparities in Expenditure Across Religions
plt.figure()
ax = sns.boxplot(data=df, x="Religion", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="Set1")
plot_title(ax, "Disparities in Expenditure Across Religions")
plt.xticks(rotation=45)
plt.show()

# 13. Outliers in Household Expenditure
plt.figure()
ax = sns.boxplot(data=df, y="Household's Usual Consumer Expenditure In A Month (Rs.)", color="coral")
plot_title(ax, "Outliers in Household Expenditure")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# General plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Helper function to add titles
def plot_title(ax, title):
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel(ax.get_xlabel(), fontsize=12)
    ax.set_ylabel(ax.get_ylabel(), fontsize=12)

# ------------------------- Demographic Analysis -------------------------

# 1. Age distribution across sectors
plt.figure()
ax = sns.histplot(data=df, x="Age", hue="Sector", bins=20, kde=True, palette="Set2")
plot_title(ax, "Age Distribution Across Sectors")
plt.show()

# 2. Population distribution by gender, religion, and social group
plt.figure()
ax = sns.countplot(data=df, x="Religion", hue="Sex", palette="muted")
plot_title(ax, "Population Distribution by Gender and Religion")
plt.xticks(rotation=45)
plt.show()

# Gender distribution by social group
plt.figure()
ax = sns.countplot(data=df, x="Social Group", hue="Sex", palette="coolwarm")
plot_title(ax, "Population Distribution by Gender and Social Group")
plt.xticks(rotation=45)
plt.show()

# 3. Average household size by sector and social group
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Household Size", ci=None, palette="husl")
plot_title(ax, "Average Household Size by Sector")
plt.show()

plt.figure()
ax = sns.barplot(data=df, x="Social Group", y="Household Size", ci=None, palette="coolwarm")
plot_title(ax, "Average Household Size by Social Group")
plt.xticks(rotation=45)
plt.show()

# 4. Marital status breakdown across age groups and genders
plt.figure()
ax = sns.countplot(data=df, x="Marital_status", hue="Sex", palette="Set3")
plot_title(ax, "Marital Status Breakdown by Gender")
plt.xticks(rotation=45)
plt.show()

plt.figure()
ax = sns.boxplot(data=df, x="Marital_status", y="Age", palette="coolwarm")
plot_title(ax, "Marital Status Across Age Groups")
plt.xticks(rotation=45)
plt.show()

# --------------------- Economic Activity Analysis ----------------------

# 5. Worker Participation Rate across sectors
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Worker Participation Ratio (CWS)", ci=None, palette="pastel")
plot_title(ax, "Worker Participation Rate Across Sectors")
plt.show()

# 6. Labor Force Participation Rate by gender and sector
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Labour Force Participation Ratio (CWS)", hue="Sex", ci=None, palette="muted")
plot_title(ax, "Labor Force Participation Rate by Gender and Sector")
plt.legend(title="Gender")
plt.show()

# 7. Employment status trends by Quarter and Month of Survey
plt.figure()
ax = sns.lineplot(data=df, x="Quarter", y="Worker Participation Ratio (CWS)", hue="Sector", marker="o", palette="Dark2")
plot_title(ax, "Trends in Worker Participation Rate by Quarter")
plt.show()

plt.figure()
ax = sns.lineplot(data=df, x="Month of Survey", y="Worker Participation Ratio (CWS)", hue="Sector", marker="o", palette="husl")
plot_title(ax, "Trends in Worker Participation Rate by Month")
plt.show()

# 8. Household type and workforce participation correlation
plt.figure()
ax = sns.boxplot(data=df, x="Household Type", y="Worker Participation Ratio (CWS)", palette="coolwarm")
plot_title(ax, "Workforce Participation by Household Type")
plt.xticks(rotation=45)
plt.show()

# --------------------- Household Expenditure Analysis ------------------

# 9. Average household consumption expenditure by sector
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Household's Usual Consumer Expenditure In A Month (Rs.)", ci=None, palette="pastel")
plot_title(ax, "Average Household Consumption Expenditure by Sector")
plt.show()

# 10. Expenditure by social group and religion
plt.figure()
ax = sns.boxplot(data=df, x="Social Group", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="husl")
plot_title(ax, "Household Expenditure by Social Group")
plt.xticks(rotation=45)
plt.show()

plt.figure()
ax = sns.boxplot(data=df, x="Religion", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="Set3")
plot_title(ax, "Household Expenditure by Religion")
plt.xticks(rotation=45)
plt.show()

# 11. Expenditure by household size
plt.figure()
ax = sns.scatterplot(data=df, x="Household Size", y="Household's Usual Consumer Expenditure In A Month (Rs.)", alpha=0.6)
plot_title(ax, "Expenditure by Household Size")
plt.show()

# --------------------- Sectoral Comparisons ----------------------------

# 12. Gender distribution in urban vs rural sectors
plt.figure()
ax = sns.countplot(data=df, x="Sector", hue="Sex", palette="coolwarm")
plot_title(ax, "Gender Distribution in Urban vs Rural Sectors")
plt.show()

# 13. WPR and LFPR comparisons across sectors
plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Worker Participation Ratio (CWS)", ci=None, palette="muted")
plot_title(ax, "Worker Participation Rate by Sector")
plt.show()

plt.figure()
ax = sns.barplot(data=df, x="Sector", y="Labour Force Participation Ratio (CWS)", ci=None, palette="Set2")
plot_title(ax, "Labor Force Participation Rate by Sector")
plt.show()

# 14. Trends in consumption expenditure over time
plt.figure()
ax = sns.lineplot(data=df, x="Month of Survey", y="Household's Usual Consumer Expenditure In A Month (Rs.)", hue="Sector", marker="o", palette="coolwarm")
plot_title(ax, "Consumption Expenditure Trends Over Time")
plt.show()

# --------------------- Relationships and Correlations ------------------

# 15. Correlation between variables
plt.figure()
corr_cols = ["Household Size", "Household's Usual Consumer Expenditure In A Month (Rs.)", 
             "Worker Participation Ratio (CWS)", "Labour Force Participation Ratio (CWS)"]
correlation_matrix = df[corr_cols].corr()
ax = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plot_title(ax, "Correlation Between Variables")
plt.show()

# --------------------- Disparity Analysis ------------------------------

# 16. Disparities in household expenditure across social groups and religions
plt.figure()
ax = sns.boxplot(data=df, x="Social Group", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="viridis")
plot_title(ax, "Expenditure Disparities Across Social Groups")
plt.xticks(rotation=45)
plt.show()

plt.figure()
ax = sns.boxplot(data=df, x="Religion", y="Household's Usual Consumer Expenditure In A Month (Rs.)", palette="coolwarm")
plot_title(ax, "Expenditure Disparities Across Religions")
plt.xticks(rotation=45)
plt.show()

# 17. Gender disparities in WPR and LFPR
plt.figure()
ax = sns.boxplot(data=df, x="Sex", y="Worker Participation Ratio (CWS)", palette="Set1")
plot_title(ax, "Gender Disparities in Worker Participation Rate")
plt.show()

plt.figure()
ax = sns.boxplot(data=df, x="Sex", y="Labour Force Participation Ratio (CWS)", palette="Set2")
plot_title(ax, "Gender Disparities in Labor Force Participation Rate")
plt.show()

# --------------------- Key Insights and Outliers -----------------------

# 18. Outliers in household expenditure
plt.figure()
ax = sns.boxplot(data=df, y="Household's Usual Consumer Expenditure In A Month (Rs.)", color="coral")
plot_title(ax, "Outliers in Household Expenditure")
plt.show()

# 19. Highest and lowest participation in workforce by group
plt.figure()
ax = sns.barplot(data=df, x="Social Group", y="Worker Participation Ratio (CWS)", ci=None, palette="coolwarm")
plot_title(ax, "Workforce Participation by Social Group")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
df = pd.read_csv("new_data.csv")

# Head of the dataset to verify the loaded data
print(df.head())

# ======= 1. Distribution of 'Industry Code (CWS)' =======
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Industry Code (CWS)', palette='viridis', order=df['Industry Code (CWS)'].value_counts().index)
plt.title("Distribution of Industry Codes (CWS)", fontsize=16)
plt.xlabel("Industry Code (CWS)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# ======= 2. Worker Participation Ratio by Sector =======
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sector', y='Worker Participation Ratio (CWS)', palette='coolwarm')
plt.title("Worker Participation Ratio by Sector", fontsize=16)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Worker Participation Ratio (CWS)", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 3. Labour Force Participation Ratio by State =======
plt.figure(figsize=(15, 8))
sns.barplot(data=df, x='State/Ut Code', y='Labour Force Participation Ratio (CWS)', palette='mako')
plt.title("Labour Force Participation Ratio by State", fontsize=16)
plt.xlabel("State/UT Code", fontsize=12)
plt.ylabel("Labour Force Participation Ratio (CWS)", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# ======= 4. Employment Trends Across Sectors =======
sector_trends = df.groupby(['Sector', 'Quarter']).size().unstack()
sector_trends.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='Set3')
plt.title("Employment Trends Across Sectors by Quarter", fontsize=16)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Number of Observations", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 5. Distribution of Vocational/Technical Training =======
plt.figure(figsize=(8, 5))
training_counts = df['Whether received any Vocational/Technical Training'].value_counts(normalize=True)
plt.figure(figsize=(8, 5))
training_counts = df['Whether received any Vocational/Technical Training'].value_counts(normalize=True)

# Ensure 'explode' matches the number of unique values
explode_values = [0.1 if i == 0 else 0 for i in range(len(training_counts))]

training_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#66c2a5', '#fc8d62'], startangle=90, explode=explode_values)
plt.title("Distribution of Vocational/Technical Training", fontsize=16)
plt.ylabel('')
plt.show()


# ======= 6. Age Distribution of Workers =======
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Age', kde=True, bins=30, color='dodgerblue')
plt.title("Age Distribution of Workers", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 7. Gender Analysis: Participation Ratio =======
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Sex', y='Worker Participation Ratio (CWS)', palette='pastel')
plt.title("Worker Participation Ratio by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Worker Participation Ratio (CWS)", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 8. Monthly Household Expenditure Analysis =======
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Household's Usual Consumer Expenditure In A Month (Rs.)", kde=True, color='coral', bins=40)
plt.title("Household's Usual Consumer Expenditure Distribution", fontsize=16)
plt.xlabel("Monthly Expenditure (Rs.)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 9. Skill Mapping Insights: Key Sectors =======
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Industry Code (CWS)', order=df['Industry Code (CWS)'].value_counts().index[:10], palette='viridis')
plt.title("Top 10 Industry Codes by Frequency", fontsize=16)
plt.xlabel("Industry Code (CWS)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout()
plt.show()

# ======= 10. Interactive Visualization for Sector vs Labour Force Participation =======
fig = px.scatter(df, x='Sector', y='Labour Force Participation Ratio (CWS)', color='State/Ut Code',
                 title="Interactive Visualization: Sector vs Labour Force Participation",
                 size='Worker Participation Ratio (CWS)', hover_data=['State/Ut Code'])
fig.show()

```
